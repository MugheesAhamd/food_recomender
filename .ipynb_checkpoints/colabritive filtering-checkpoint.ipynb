```python

import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python

ratings = pd.read_csv('food_ratings.csv')#opeing food_ratings with pandas
recipes_names = pd.read_csv('all_recipes.csv').drop_duplicates(subset='title') # opening all the recipes and droping duplicated values and titles
recipes_and_ratings = pd.merge(ratings , recipes_names , on='recipe_id')#merging food ratings and recipes
```

```python
recipes_names.head()
```

```python
recipes_and_ratings.head()
```

cleaning data with deleting users with less than 4 ratings

```python
better_user_ratings = recipes_and_ratings.pivot_table(index='title',columns='user_id',values='rating')#pivoting tables setting title as index and user_idies as columns
better_user_ratings = better_user_ratings.dropna(thresh=4 , axis=1)#droping nan vlues users with threshold of 4
print(recipes_and_ratings.shape)

better_user_ratings.head()
```

replacing new ids with old ids and deleteing ratings with no user id

```python
new_user_list = better_user_ratings.columns.tolist()
recipes_and_ratings = recipes_and_ratings.loc[recipes_and_ratings['user_id'].isin(new_user_list)]
print(recipes_and_ratings.shape)
recipes_and_ratings.head()
```

```python
recipes_titles= recipes_names['title'].str.replace('"','').tolist()#replacing " to empty space for avoiding errors in api
```

```python
recipes_and_ratings =recipes_and_ratings.pivot_table(index='user_id', columns='title', values='rating')
recipes_and_ratings.head()
```

```python
recipes_and_ratings = recipes_and_ratings.fillna(0)
recipes_and_ratings.head()
```

```python
def standarize(row):
    new_row = (row -row.mean()) / (row.max()-row.min())
    return new_row
recipes_and_ratings = recipes_and_ratings.apply(standarize)
recipes_and_ratings.head()
```

```python
recipes_and_ratings.shape
```

```python
recipes_and_ratings = recipes_and_ratings.iloc[0:1151,:]
```

calculating cosine similarity and turning it to a dataframe to be able to save it on disk

```python

item_similarity = cosine_similarity(recipes_and_ratings)
item_similarity = pd.DataFrame(item_similarity,columns=recipes_titles,index=recipes_titles)
item_similarity.head()
```

testing the model

```python
food_name = "Chef John's Carrot Cake"
user_rating = 5
similar_score = item_similarity[food_name]*(user_rating-2.5)
similar_score = similar_score.sort_values(ascending=False)
print(similar_score)
```
